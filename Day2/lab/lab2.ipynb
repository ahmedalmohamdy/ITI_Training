```python
from re import X

x=int(input("please enter any number : "))

sum=0
for i in range (x+1):
  sum+=i
print("the sum of natural numbers from 1 to "+str(x)+"= "+str(sum))
average = sum/x
print("the average of natural numbers from 1 to "+ str(x) +"= "+str(average))
```
